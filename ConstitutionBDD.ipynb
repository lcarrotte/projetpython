### Base d'infos sur les députés


```python
import ssl
import numpy as np
import requests
import pandas as pd
import zipfile
from io import BytesIO
ssl._create_default_https_context = ssl._create_unverified_context
```

```python

url = 'https://data.assemblee-nationale.fr/static/openData/repository/15/amo/deputes_actifs_mandats_actifs_organes_divises/AMO40_deputes_actifs_mandats_actifs_organes_divises_XV.csv.zip'
print("debut téléchargement")
req = requests.get(url)
zf= zipfile.ZipFile(BytesIO(req.content))
print("fin téléchargement")
names=zf.namelist()
print("début extraction") #Il y a un fichier par député donc on fait la fusion
for name in names:
    if name[:6]=="acteur": #les autres fichiers ne sont pas intéressants
        df = pd.read_csv(zf.open(name),";").loc[:,['etatCivil[1]/ident[1]/prenom[1]', 'etatCivil[1]/ident[1]/nom[1]','etatCivil[1]/infoNaissance[1]/villeNais[1]','etatCivil[1]/infoNaissance[1]/depNais[1]','profession[1]/libelleCourant[1]','profession[1]/socProcINSEE[1]/catSocPro[1]','uri_hatvp[1]']]
        try:
            tab=tab.append(df)
        except:
            tab=df
            
```

```python
tab2=pd.read_csv("https://www.nosdeputes.fr/deputes/csv",";")
tab2=tab2.drop(['id','lieu_naissance', 'sites_web', 'emails', 
       'collaborateurs','anciens_mandats', 'profession', 'place_en_hemicycle', 'adresses','url_an',
       'id_an', 'slug', 'url_nosdeputes', 'url_nosdeputes_api'],axis=1)
```

```python
df=tab2.merge(tab,right_on=['etatCivil[1]/ident[1]/nom[1]','etatCivil[1]/ident[1]/prenom[1]'], left_on=['nom_de_famille','prenom'])
df=df.drop(['etatCivil[1]/ident[1]/prenom[1]', 'etatCivil[1]/ident[1]/nom[1]'],axis=1)
df=df.rename(columns={'etatCivil[1]/infoNaissance[1]/villeNais[1]': 'ville_naissance', 'etatCivil[1]/infoNaissance[1]/depNais[1]': 'dep_naissance','profession[1]/libelleCourant[1]':'profession','profession[1]/socProcINSEE[1]/catSocPro[1]':'cat_soc_pro','uri_hatvp[1]':'patrimoine'})
df['ancrage_local']=np.where( df['nom_circo'] == df['dep_naissance'] , '1', '0')
df.head(10)
```

```python
df['sexe']=[1 if s=='H' else 0 for s in df['sexe']]
```

```python
df['num_deptmt']=[s if len(s)==2 else "0"+s for s in df["num_deptmt"]]
df['num_circo']=[s if len(str(s))==2 else "0"+str(s) for s in df["num_circo"]]
df["DEPSIR"]=df['num_deptmt'].astype(str)+df['num_circo'].astype(str)
```

```python
df.to_csv("caract_deputes.csv",";",index=False)
```
