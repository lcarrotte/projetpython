### ACP sur l'implication à l'AN 

```python
data_pca=tab_final[['semaines_presence/duree_mandat (sem)','amendements_adoptes/duree_mandat (sem)','amendements_proposes/duree_mandat (sem)','amendements_signes/duree_mandat (sem)','commission_interventions/duree_mandat (sem)','commission_presences/duree_mandat (sem)','hemicycle_interventions/duree_mandat (sem)','hemicycle_interventions_courtes/duree_mandat (sem)','propositions_ecrites/duree_mandat (sem)','propositions_signees/duree_mandat (sem)','questions_ecrites/duree_mandat (sem)','questions_orales/duree_mandat (sem)','rapports/duree_mandat (sem)']]
data_pca.head(10)
```

```python
import numpy as np
from sklearn import decomposition
from sklearn import preprocessing

n_comp = 2

X = data_pca.values
features = data_pca.columns

# Centrage et Réduction
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)

# Calcul des composantes principales
pca = decomposition.PCA(n_components=n_comp)
pca.fit(X_scaled)


pcs = pca.components_

# Projection des individus
X_projected = pca.transform(X_scaled)
X=[]
Y=[]
for i in range(len(X_projected)):
    X.append(X_projected[i][0])
    Y.append(X_projected[i][1])
data_pca["X"]=X
data_pca["Y"]=Y
```

```python
import plotly.express as px
px.scatter(data_pca, x="X", y="Y",text=data_pca.index,title="ACP de l'implication à l'AN")
```

```python
from matplotlib import pyplot as plt
import matplotlib.colors as mcolors
x=[]
y=[]
for i in range(len(pca.components_[0])):
    x.append(pca.components_[0][i])
    y.append(pca.components_[1][i])

fig = plt.figure(figsize=(8,8),dpi=100)
ax = fig.add_subplot(111)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
cercle = plt.Circle((0, 0), 1, color='b', fill=False)
ax.add_artist(cercle)
colors=list(mcolors.TABLEAU_COLORS.values())+['red','blue','black','grey']
for j in range(len(x)):
    arrow=ax.arrow(0, 0, x[j], y[j], head_width=0.02, head_length=0.1,label=data_pca.columns[j],ec=colors[j],fc=colors[j])
    plt.legend()
plt.title("Cercle des corrélations de l'ACP",size=17)
plt.savefig("ACP_cercle.png")
```

```python
print(pca.explained_variance_ratio_)
```
